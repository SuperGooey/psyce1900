### The data you started with were in a csv file.  What file format does your program save data in?

R can save a dataframe to the working directory as a CSV with the command:

`write.csv(Data,"my_data.csv")`

If we need to save to a different local directory, we can use the double backslash \\ to "escape" out of the single backslash command in R and instead save the file to a specific file path, like:

`write.csv(Data, "C:\\Users\\Colin\\")`

Additionally, we can also save the R script file via the File menu in RStudio.
